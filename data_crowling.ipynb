```python
import csv
import re
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import time
```

```python
players = [
'최형우', '장성우', '이원석', '박병호', '문보경', '소크라테스', '마티니', '터크먼', '노진혁', '푸이그',
'류지혁', '이지영', '박민우', '강승호', '정수빈', '황대인', '유강남', '송성문', '김태연', '에레디아',
'오스틴', '김민혁', '문성주', '알포드', '마틴', '김현준', '문현빈', '김민석', '로하스', '이재현',
'김주원', '레이예스', '김도영', '김지찬', '나승엽', '고승민', '데이비슨', '권희동', '신민재', '윤동희',
'이우성', '서호철', '페라자', '이주형', '박승욱', '김휘집', '김영웅'
]

# 드라이버 세팅
options = webdriver.ChromeOptions()
# options.add_argument("--headless=new") 
options.add_argument('--start-maximized')
driver = webdriver.Chrome(options=options)
wait = WebDriverWait(driver, 10)

driver.get("https://statiz.sporki.com/player/?m=playerinfo")
time.sleep(1)

results = []

for name in players:
    try:
        # 검색 버튼 클릭
        wait.until(EC.element_to_be_clickable((By.CLASS_NAME, 'btn_box'))).click()

        # 검색창에 이름 입력
        search_input = wait.until(EC.presence_of_element_located((By.ID, 's')))
        search_input.clear()
        search_input.send_keys(name)

        # 검색
        wait.until(EC.element_to_be_clickable((By.CSS_SELECTOR, "input.btn"))).click()

        wait.until(EC.presence_of_element_located((By.CLASS_NAME, 'p_info')))
        wait.until(EC.presence_of_element_located((By.CLASS_NAME, 'man_info')))

        # 팀명, 투타
        span_elems = driver.find_elements(By.CSS_SELECTOR, '.p_info .con span')
        team = span_elems[0].text if len(span_elems) > 0 else ""
        handedness = span_elems[-1].text if len(span_elems) > 1 else ""

        # 생년월일 추출 및 정제
        birth_date_raw = ""
        li_elems = driver.find_elements(By.CSS_SELECTOR, '.man_info li')
        for li in li_elems:
            if "생년월일" in li.text:
                birth_date_raw = li.text.split("：")[-1].strip()
                break

        # 정제: YYYY-MM-DD 형식으로 정확히 추출
        match = re.search(r'(\d{4})년\s*(\d{2})월\s*(\d{2})일', birth_date_raw)
        if match:
            birth_date_clean = f"{match.group(1)}-{match.group(2)}-{match.group(3)}"
        else:
            birth_date_clean = birth_date_raw

        results.append({
            "이름": name,
            "팀명": team,
            "투타": handedness,
            "생년월일": birth_date_clean
        })

        print(f"{name} 완료")
        time.sleep(1)

    except Exception as e:
        print(f"{name} 오류: {e}")

driver.quit()

# CSV 저장
with open("player_info4.csv", "w", newline="", encoding="cp949") as f:
    writer = csv.DictWriter(f, fieldnames=["이름", "팀명", "투타", "생년월일"])
    writer.writeheader()
    writer.writerows(results)
```
