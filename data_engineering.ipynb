```python
unique_players = df['선수명'].unique()
print(unique_players)
```

```python
player_year_counts = df.groupby('선수명')['연도'].nunique()

# 4년 모두 등장한 선수만 필터링
players_all_years = player_year_counts[player_year_counts == 4].index.tolist()
print(f"2021~2024년 모두 출장한 선수 수: {len(players_all_years)}명")
```

```python
df_cleaned = df.drop(columns=['순위.1', '순위.2', '선수명.1', '선수명.2', '팀명.1', '팀명.2', 'AVG.1', 'AVG.2'])
```

```python
df_cleaned.to_csv('./안쓰는데이터/kbo_21_24_cleaned.csv', index=False)
```
