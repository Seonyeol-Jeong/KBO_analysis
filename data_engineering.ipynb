```python
df = pd.read_csv('./안쓰는데이터/kbo_18_20_cleaned.csv')
df.head()
```

```python
df.columns
```

```python
df_팀성적 = pd.read_csv('./안쓰는데이터/kbo_18_20_팀순위.csv')
df_팀성적
```

```python
df_merged2 = pd.merge(df, df_팀성적, on=['연도', '팀명'], how='left')
df_merged2.columns
```

```python
# 1. _y 접미사 붙은 컬럼만 남기고 이름 원복
df_merged2.rename(columns={
    '팀순위_y': '팀순위',
    '경기수_y': '경기수',
    '팀승수_y': '팀승수',
    '팀패수_y': '팀패수',
    '팀무승부수_y': '팀무승부수',
    '팀승률_y': '팀승률',
    '팀게임차_y': '팀게임차'
}, inplace=True)
# df_merged2.columns
# 2. _x 접미사 붙은 컬럼 제거
df_merge3=df_merged2.drop(columns=[
    '팀순위_x', '경기수_x', '팀승수_x', '팀패수_x',
    '팀무승부수_x', '팀승률_x', '팀게임차_x'])

# 3. 확인
print(df_merge3.columns)
```

```python
## 포지션, 투구, 타격, 나이
```

```python
# 중복되는 선수 출생연도, 투구유형, 타격유형, 포지션 병합
df_21 = pd.read_csv('./kbo_21_24_merged.csv',encoding='cp949')

# 1. 사용할 컬럼만 추출 (중복 제거 포함)
player_info = df_21[['선수명', '출생연도', '투구유형', '타격유형', '포지션']].drop_duplicates()

# 2. 선수명 기준으로 df_merge3에 병합
df_merge4 = pd.merge(df_merge3, player_info, on='선수명', how='left')
df_merge4.columns

# 3. _y 컬럼을 본래 컬럼명으로 바꾸기
df_merge4.rename(columns={
    '출생연도_y': '출생연도',
    '투구유형_y': '투구유형',
    '타격유형_y': '타격유형',
    '포지션_y': '포지션'
}, inplace=True)

# 4. _x 컬럼 제거
df_merge5 = df_merge4.drop(columns=['출생연도_x', '투구유형_x', '타격유형_x', '포지션_x'])

# 5. 결과 확인
print(df_merge5[['선수명', '출생연도', '투구유형', '타격유형', '포지션']].head())
```

```python
df_merge5
```

```python
df_merge5.to_csv('./안쓰는데이터/kbo_18_20_merged.csv', index=False)
```

```python
df = pd.read_csv('./안쓰는데이터/kbo_18_20_merged.csv')
df.head()
```

```python
# 결측값이 있는 선수명
missing_players = df_merge5[df_merge5[['출생연도', '투구유형', '타격유형', '포지션']].isna().any(axis=1)]
print(missing_players['선수명'].unique())
```

```python
df_info = pd.read_csv('./안쓰는데이터/player_info3.csv',encoding='cp949')
df_info.head()
```

```python
# df와 df_info 조합: 선수명 기준으로 포지션 투구유형 타격유형 출생연도 추가 (NAN값이면)

player_info = df_info[['선수명', '출생연도', '투구유형', '타격유형', '포지션']].drop_duplicates()
df_final_18_20 = pd.merge(df, player_info, on='선수명', how='left', suffixes=('', '_info'))

# 대체
for col in ['출생연도', '투구유형', '타격유형', '포지션']:
    df_final_18_20[col] = df_final_18_20[col].fillna(df_final_18_20[f"{col}_info"])
df_final_18_20.drop(columns=[f'{col}_info' for col in ['출생연도', '투구유형', '타격유형', '포지션']], inplace=True)

print(df_final_18_20[['선수명', '출생연도', '투구유형', '타격유형', '포지션']].head())
```

```python
df_final_18_20.isna().sum()
```

```python
df_final_18_20.to_csv('./kbo_18_20_merged.csv', index=False)
```

## 최종 데이터 병합

```python
df1 = pd.read_csv('./kbo_18_20_merged.csv')
df2 = pd.read_csv('./kbo_21_24_merged.csv',encoding='cp949')
```

```python
print(df1.columns)
print(df2.columns)
```

```python
# df1 기준으로 df2의 열 순서 정렬
df2 = df2[df1.columns]

# 병합 (행 방향)
df_all = pd.concat([df1, df2], axis=0, ignore_index=True)

# 연도 기준 정렬
df_all.sort_values(by='연도', inplace=True)
df_all.reset_index(drop=True, inplace=True)

# 결과 확인
print(df_all.shape)
print(df_all.columns)
```

```python
df_all.isna().sum()
```

```python
df_all.to_csv('./kbo_18_24_merged.csv', encoding='cp949', index=False)
```
