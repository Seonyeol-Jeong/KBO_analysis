```python
# 팀순위와의 상관계수 정렬 (절댓값 기준)
ops_corr = corr_matrix['팀순위']
ops_corr_sorted = ops_corr.reindex(ops_corr.abs().sort_values(ascending=False).index)

ops_corr_df = ops_corr_sorted.to_frame().T
ops_corr_df.index = ['팀순위와의 상관계수']
ops_corr_df.round(2)
```

상관계수 수치로 top 20 보기

```python
# 상관관계 행렬 재계산 (불필요한 반복 피함)
corr_matrix = numeric_df.corr()

# 상관계수 절댓값 기준으로 정렬하기 위해 행렬을 풀어서 시리즈로 변환
corr_pairs = corr_matrix.unstack()

# 자기 자신과의 상관 제외하고 절댓값 기준 상위 n개 추출
corr_pairs = corr_pairs[corr_pairs.index.get_level_values(0) != corr_pairs.index.get_level_values(1)]
top_corr = corr_pairs.abs().sort_values(ascending=False).drop_duplicates().head(20)

# 실제 상관계수 값으로 재정렬
top_corr_values = corr_pairs[top_corr.index]

# 데이터프레임으로 정리
top_corr_df = pd.DataFrame({
    '변수1': [i[0] for i in top_corr_values.index],
    '변수2': [i[1] for i in top_corr_values.index],
    '상관계수': top_corr_values.values
})

top_corr_df
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# 연도를 문자열로 처리하여 x축을 범주형으로 고정
df['연도_str'] = df['연도'].astype(str)
yearly_rank = df.groupby(['팀명', '연도_str'])['팀순위'].mean().round(2).reset_index()

# FacetGrid로 박스플롯 생성
g = sns.FacetGrid(df, col='팀명', col_wrap=5, height=4, sharey=True)
g.map_dataframe(sns.boxplot, x='연도_str', y='OBP')

# 박스 위에 순위 텍스트 표시
for ax in g.axes.flat:
    title = ax.get_title()
    team = title.split('=')[-1].strip()
    ax.set_title(f"{team}")
    
    team_data = yearly_rank[yearly_rank['팀명'] == team]
    for i, row in team_data.iterrows():
        ax.text(x=row['연도_str'], y=0.455, s=f"{row['팀순위']}", 
                ha='center', fontsize=9, color='black')

# 전체 타이틀
g.fig.subplots_adjust(top=0.9)
g.fig.suptitle('팀별 시즌별 OBP Boxplot (박스 위 숫자는 평균 순위)', fontsize=16)
plt.tight_layout()
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# 팀별 OPS 분포 시각화 (히스토그램)
g = sns.FacetGrid(df, col='팀명', col_wrap=5, height=4, sharex=True, sharey=True)
g.map_dataframe(sns.histplot, x='OPS', bins=30, kde=True)
g.set_titles(col_template="{col_name}")
g.fig.subplots_adjust(top=0.9)
g.fig.suptitle('팀별 OPS 분포 (히스토그램)', fontsize=16)
plt.tight_layout()
plt.show()
```

제목에 평균순위 추가

```python
import seaborn as sns
import matplotlib.pyplot as plt

# 팀별 평균 순위 계산
team_rank_mean = df.groupby('팀명')['팀순위'].mean().round(2)

# 팀별 OPS 분포 시각화 (히스토그램)
g = sns.FacetGrid(df, col='팀명', col_wrap=5, height=4, sharex=True, sharey=True)
g.map_dataframe(sns.histplot, x='OPS', bins=30, kde=True)

# 제목에 평균 순위 추가 (팀명 정확히 추출)
for ax in g.axes.flat:
    full_title = ax.get_title()                  # 예: "팀명 = 키움"
    team_name = full_title.split('=')[-1].strip()  # "키움"
    rank = team_rank_mean.get(team_name, "N/A")
    ax.set_title(f"{team_name} (평균순위: {rank})")

# 전체 제목 및 레이아웃
g.fig.subplots_adjust(top=0.9)
g.fig.suptitle('팀별 OPS 분포 (히스토그램)', fontsize=16)
plt.tight_layout()
plt.show()
```

성적 카테고리별 구분 출력

```python
import seaborn as sns
import matplotlib.pyplot as plt

# 팀 그룹 분류
group_high = ['NC', '롯데', 'SSG']                     # OPS 높음
group_balanced = ['LG', '두산', 'KIA']                 # 중간~균형
group_skewed = ['삼성', '키움']                        # 편차 큼 (극단치 있음)
group_low = ['KT', '한화']                             # OPS 낮음

# 팀별 평균 순위 계산
team_rank_mean = df.groupby('팀명')['팀순위'].mean().round(2)

# 그룹별 시각화 함수
def plot_group(title, team_list):
    g = sns.FacetGrid(df[df['팀명'].isin(team_list)], 
                      col='팀명', col_wrap=5, height=4, sharex=True, sharey=True)
    g.map_dataframe(sns.histplot, x='OPS', bins=30, kde=True)
    g.set_titles(col_template="{col_name}")
    g.fig.subplots_adjust(top=0.85)
    g.fig.suptitle(f"[{title}] 팀별 OPS 분포 (히스토그램)", fontsize=14)
    plt.tight_layout()
    plt.show()

# 각 그룹 출력
plot_group("OPS 높은 팀", group_high)
plot_group("OPS 균형 분포 팀", group_balanced)
plot_group("OPS 편차 큰 팀", group_skewed)
plot_group("OPS 낮은 팀", group_low)
```
